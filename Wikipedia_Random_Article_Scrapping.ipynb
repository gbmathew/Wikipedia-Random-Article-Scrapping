```python
import requests
from bs4 import BeautifulSoup
from datetime import date
```

```python
today = date.today()
today = str(today)
print(today)
```

```python
fileObject = open('article_content.txt', "w", encoding="utf-8")
fileObject.write('Article_Extract_Date : ')
fileObject.write(today)
fileObject.write("\n")
fileObject.write("\n")
fileObject.close()
```

```python
print("Total number of articles to extract : ")
num = input()
num = int(num)
num = num+1
```

```python
i = 1
while i < num:
    url = requests.get("https://en.wikipedia.org/wiki/Special:Random")
    soup = BeautifulSoup(url.content, 'html5lib')
    title = soup.find(class_ ="vector-body").text
    url = "https://en.wikipedia.org/wiki/%s" % title
    text = soup.find('p').getText()
    if text is not None:
        with open('article_content.txt', "a", encoding="utf-8") as f:
            f.write(title)
            f.write("\n")
            f.write(text)
            f.write("\n")
            print(i, " Completed")
    else: 
        print(i, "Error Scrapping")
    
    i = i+1

with open('article_content.txt', "a", encoding="utf-8") as f:
    f.close()
```
